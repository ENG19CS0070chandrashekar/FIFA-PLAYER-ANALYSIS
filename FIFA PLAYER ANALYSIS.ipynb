# univariate bivariate and multivariate analysis

### Univariate Analysis

### Bar Chart

```python
age = fifa[['short_name','age','wage_eur']]
```

```python
age
```

```python
top_salary = age.sort_values(by = ['wage_eur'],ascending = False)
```

```python
top_salary
```

```python
plt.figure(figsize = (15,8))
barlist = plt.bar(list(top_salary['short_name'].head(7)),list(top_salary['age'].head(7)))
plt.title("Player Age",fontsize = 15)
plt.xlabel('Name',fontsize = 12)
plt.ylabel('Age',fontsize = 12)
barlist[2].set_color("g")
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.barh(list(fifa['nationality'].value_counts()[:10].keys()),list(fifa['nationality'].value_counts()[:10]),color = 'g')
plt.title("Nationality of players",fontsize = 15)
plt.xlabel("Nationality",fontsize = 13)
plt.ylabel("Number of players",fontsize = 13)
plt.show()
```
